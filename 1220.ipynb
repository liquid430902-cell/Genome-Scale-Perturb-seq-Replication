```python
def extract_gene_name_from_index(index_str):
    """从索引字符串中提取基因名"""
    # 格式: '10023_ZC3H18_P1P2_ENSG00000158545'
    parts = index_str.split('_')
    if len(parts) >= 2:
        return parts[1]  # 基因名是第二部分
    else:
        return index_str
```
